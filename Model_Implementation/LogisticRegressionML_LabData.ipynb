```python
import pandas as pd
# encoding of categorical columns
df_encoded = pd.get_dummies(df, columns=['city', 'occupation'], drop_first=True)

df_encoded.head(5)
```

```python
# Model Implementation
from sklearn.linear_model import LinearRegression
from sklearn.metrics import classification_report, accuracy_score
from sklearn.model_selection import train_test_split
x = df_encoded.drop(columns=['insurance_premium_category'])
y = df['insurance_premium_category']   # keep as labels for classification
```

```python
x.head(4)
```

```python
y.head(4)
```

```python
y = pd.get_dummies(y, columns=['insurance_premium_category'], drop_first=True)
```

```python
y.head(4)
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42, stratify=y
)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=1000, multi_class='multinomial')
model.fit(x_train, y_train)

predict = model.predict(x_test)
print("Accuracy:", accuracy_score(y_test, predict))
print("\nClassification Report:\n", classification_report(y_test, predict))
```
